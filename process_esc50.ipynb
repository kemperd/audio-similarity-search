```python
import pandas as pd
import shutil
import os
```

```python
df = pd.read_csv('ESC-50-master/meta/esc50.csv')
```

```python
df.groupby(['category'])['category'].count()
```

```python
df[df['category'] == 'crow']
```

```python
SOURCE_ROOT = 'ESC-50-master/audio'
TARGET_ROOT = 'esc50'

TRAIN_DIR = TARGET_ROOT + '/train' 
TEST_DIR = TARGET_ROOT + '/test'
VAL_DIR = TARGET_ROOT + '/val'

if not os.path.exists(TRAIN_DIR): os.makedirs(TRAIN_DIR)
if not os.path.exists(TEST_DIR): os.makedirs(TEST_DIR)

for index, item in df.iterrows():
    class_train_dir = TRAIN_DIR + '/' + item['category']
    class_test_dir = TEST_DIR + '/' + item['category']
    class_val_dir = VAL_DIR + '/' + item['category']

    if not os.path.exists(class_train_dir): os.makedirs(class_train_dir)
    if not os.path.exists(class_test_dir): os.makedirs(class_test_dir)
    if not os.path.exists(class_val_dir): os.makedirs(class_val_dir)

    if len(os.listdir(class_train_dir)) < 34:
        target_dir = class_train_dir
    elif len(os.listdir(class_train_dir)) >= 34 and len(os.listdir(class_test_dir)) < 5:
        target_dir = class_test_dir
    elif len(os.listdir(class_test_dir)) >= 5 and len(os.listdir(class_val_dir)) < 1:
        target_dir = class_val_dir

    if not os.path.exists(target_dir): os.makedirs(target_dir)

    file_source = SOURCE_ROOT + '/' + item['filename']
    file_target = target_dir + '/' + item['filename']

    shutil.copy(file_source, file_target)
        
```
